```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
swiggy = pd.read_csv(r'C:\Users\admin\Downloads\Swiggy Restaurants dataset.csv')
```

```python
swiggy.head()
```

```python
swiggy.info()
```

Q1. Which are the Top 10 restaurants with highest average ratings?

```python
top_rated = swiggy.sort_values(by = 'Avg ratings',ascending=False).head()
```

```python
top_rated
```

```python
plt.figure(figsize=(10,5))
plt.bar(top_rated['Restaurant'],top_rated['Avg ratings'],color = 'skyblue')
plt.xlabel('Avg Rating')
plt.title('Top 10 Restaurant by Average Ratings')
plt.show()
```

Q2. Which cities have the most restaurants listed?

```python
city_counts = swiggy['City'].value_counts().head()
```

```python
city_counts
```

```python
plt.figure(figsize=(10,5))
city_counts.plot(kind="bar", color="lightgreen")
plt.ylabel("Number of Restaurants")
plt.title("Top 10 Cities with Most Restaurants")
plt.show()
```

Q3. What are the most common cuisines (Food type)?

```python
food_exploded = swiggy["Food type"].str.split(",").explode()
top_foods = food_exploded.value_counts().head(10)
```

```python
top_foods
```

```python
plt.figure(figsize=(10,5))
top_foods.plot(kind="bar", color="green")
plt.ylabel("Count")
plt.title("Top 10 Most Common Food Types")
plt.show()
```

Q4. Distribution of delivery times

```python
plt.figure(figsize=(8,5))
plt.hist(swiggy["Delivery time"], bins=20, edgecolor="black")
plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Number of Restaurants")
plt.title("Distribution of Delivery Times")
plt.show()
```

```python
swiggy.sort_values(by = 'Delivery time',ascending = False).head()
```

```python
swiggy.sort_values(by = 'Delivery time').head()
```

Q5. Relationship between Price and Ratings

```python
plt.figure(figsize=(10,5))
plt.scatter(swiggy['Price'],swiggy['Avg ratings'],alpha=0.5)
plt.xlabel("Price (₹)")
plt.ylabel("Average Rating")
plt.title("Price vs Average Rating")
plt.show()
```

Q6. Which are the cheapest restaurants (lowest price)?

```python
cheapest = swiggy.sort_values(by="Price", ascending=True).head(10)
```

```python
cheapest
```

```python
plt.figure(figsize=(10,5))
plt.bar(cheapest["Restaurant"], cheapest["Price"], color="teal")
plt.xlabel("Price (₹)")
plt.ylabel("Restaurant")
plt.title("Top 10 Cheapest Restaurants")
plt.show()
```

Q7. Which are the fastest delivery restaurants?

```python
fastest = swiggy.sort_values(by="Delivery time", ascending=True).head(10)
```

```python
fastest
```

```python
plt.figure(figsize=(10,5))
plt.barh(fastest["Restaurant"], fastest["Delivery time"], color="green")
plt.xlabel("Delivery Time (minutes)")
plt.title("Top 10 Fastest Delivery Restaurants")
plt.show()
```

Q8. Which cities have the fastest average delivery times?

```python
city_delivery = swiggy.groupby("City")["Delivery time"].mean().sort_values().head(10)
```

```python
city_delivery
```

```python
plt.figure(figsize=(10,5))
city_delivery.plot(kind="bar", figsize=(10,5), color="skyblue")
plt.ylabel("Average Delivery Time (minutes)")
plt.title("Top 10 Cities with Fastest Deliveries")
plt.show()
```
