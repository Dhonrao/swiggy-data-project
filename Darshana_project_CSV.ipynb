Plot Transactions by Payment Method (Card vs Cash)

```python
plt.figure(figsize=(6,6))
plt.pie(spend_counts, labels=spend_counts.index, autopct="%1.2f%%",startangle=90)
plt.title("Transactions by Payment Method")
plt.show()
```

Plot Customer Age Distribution

```python
plt.figure(figsize=(8,5))
plt.hist(highest_age["'age'"], bins=10, edgecolor="black")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.title("Customer Age Distribution")
plt.show()
```

Plot Top 5 Customers by Spending

```python
top_customers = (
    df_transactions.groupby("'custno'")["'amount'"].sum().nlargest(5).reset_index()
)
top_customers = top_customers.merge(df_customers, on="'custno'")

plt.figure(figsize=(8,5))
plt.bar(top_customers["'firstname'"] + " " + top_customers["'lastname'"], top_customers["'amount'"])
plt.xlabel("Customer")
plt.ylabel("Total Spending")
plt.title("Top 5 Customers by Spending")
plt.show()
```

```python
top_products = (
    df_transactions["'productno'"]
    .value_counts()
    .head(5)
    .reset_index()
)
top_products.columns = ["'productno'", "'count'"]

# Merge with product names
top_products = top_products.merge(df_products, on="'productno'")

# Plot bar chart
plt.figure(figsize=(8,5))
plt.bar(top_products["'productName'"], top_products["'count'"])
plt.xlabel("Product Name")
plt.ylabel("Number of Purchases")
plt.title("Top 5 Most Demanded Products")
plt.show()
```
